```python
# K-Fold Cross-Validation
from sklearn import datasets, linear_model
from sklearn.model_selection import cross_val_score
diabetes = datasets.load_diabetes()
X = diabetes.data[:150]
y = diabetes.target[:150]
lasso = linear_model.Lasso()
print(cross_val_score(lasso, X, y, cv=3))
```

Bias: is the error between average model prediction and the ground truth, happen due to incorrect assumptions in ML process.
Variance: refer to changed in model when using different portions of training data set

there is relationship between Bias and Variance:
1- high Bias and low Variance: model is too simple and does not capture trend of data (toom uch accure)
2- low Bias and high Variance: model is too sensitive and is capturing noise (overfitting) (toom uch accure)
3- high Bias and high Variance: the model is too simple to capture the data trends and too sensitive and capture noise
4- low Bias and low Variance: Great model, capturing the trends of data and generalized well to unseen data (rare accure)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_excel(r'E:\AI Projects\ML Course\ReinforcementLearning\Ads.xlsx')
x = dataset.iloc[:, [4,5]].values
y = dataset.iloc[:, -1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(x_train, y_train)

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X= x_train, y = y_train, cv = 3 ) 
print(accuracies.mean())
print(accuracies.std())
```
