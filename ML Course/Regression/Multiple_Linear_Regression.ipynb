Assumptions of Linear Regression
1 - Linearity: linear relationship between x , y
2 - Homoscedasticity: equal variance in observations
3 - Multi-variate Normality: Normality of error distribution
4 - Independence: no auto-correlation in observwtions
5 - Lack of Multicollinearity: predicators (x1, x2, ...) are nor correlated with each other

P-Value and Hypothesis
Coin throwing, if the H0 is the fair coin:
1th throw feeling of line: 0.5
2th throw feeling of line: 0.25
3th throw line: 0.12
4th throw line: 0.06
5th throw line: 0.03
if all 5 throw is line you doubt about the fairness of coin. 
so we set p-value = 0.05, means if the posibility go under the 0.05 it means the H0 is rejected with 95% confidence

Methods of building models:
1 - All-in
2 - Backward Elimination (remove variable when P-value > significance level)
3 - Forward Selection (build simple regression model with all predictors and select the one with lowest p-value
                       keep the variable and add extra predictors, consider the predictor with lowest p-value
                       if p-value < significance level add another predictor till p-value > significance)
4 - Bidirectional Elimination ( combination of two model above)
5 - Score Comparison

In multiple linear regression we don't need feature scaling, because coefficients compensate unbalance in data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('dataset.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)) #axis=1 combines row by row
```

```python
# Evaluation of model performance
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
