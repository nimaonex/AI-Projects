```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# iloc: locate indexes, it take the indexes of the columns and rows we want to extract
dataset = pd.read_excel('MLTest.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan, strategy= 'mean')
imputer.fit(x[:, 3:5])
x[:, 3:5] = imputer.transform(x[:, 3:5])
```

```python
# for better analysis and escape from meaningful numbering for categorical data, with need some encoding to prevent correlation for ML
# it made columns for distinct number of categories
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
# Lable encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# splitting the dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1) 
# test size means the percent of main data considered for test set
# random state, to randomly split data
```

```python
# Feature Scaling, it is not nessecary to do in each ML project
# Standardization: x - mean(x) / sd(x)  --> (-3,3) recommendded more to use
# Normalization: x - min(x) / max(x) - min(x) --> (0,1) use in special times
# It should not apply on dummy variables such as encoders
# for train data set fit and transform for calculation in which fit calculate mean, stdv and transform clculate total formula
# for test data set, we just use transform object and use same scaling used for training set to test set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 4:] = sc.fit_transform(x_train[:, 4:])
x_test[:, 4:] = sc.transform(x_test[:, 4:])
```
