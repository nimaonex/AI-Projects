```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r'E:\AI Projects\ML Course\Regression\Reg_Files\salary.xlsx')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
y
```

we don't need feature scaling for desicion tree model

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0) # random_state allow us to get same result at the end
regressor.fit(x, y)
```

```python
# salary dataset is not appropriate for desicion tree model
x_grid = np.arange(min(min(x)), max(max(x)), 0.1) # consider this part of code for better solution
x_grid = x_grid.reshape((len(x_grid)), 1)
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Tree')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```
