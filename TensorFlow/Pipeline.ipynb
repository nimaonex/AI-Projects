```python
sample = tf.random.uniform([200, 10])
tf_s_dataset = tf.data.Dataset.from_tensor_slices(sample)

for r, c in enumerate(tf_s_dataset):
    print(c)
    if r > 10:
        break
```

```python
sample_x = tf.random.uniform([200, 10])
sample_y = tf.random.uniform([200], maxval=2, dtype=tf.int32)

tf_dataset = tf.data.Dataset.from_tensor_slices((sample_x, sample_y))
# OR
tf_dataset_x = tf.data.Dataset.from_tensor_slices(sample_x)
tf_dataset_y = tf.data.Dataset.from_tensor_slices(sample_y)
tf_dataset_zip = tf.data.Dataset.zip(tf_dataset_x, tf_dataset_y)


for i, (x,y) in enumerate(tf_dataset):
    print(f"x is: {x}")
    print(f"y is: {y}")
    if i > 5:
        break
# OR

for x, y in tf_dataset.take(10): 
    print(x)
    print(y)
```

```python
range = tf.range(5)
dataset_range = tf.data.Dataset.from_tensor_slices(range)
_shuffle = dataset_range.shuffle(buffer_size=5) # print the result each time without any order

for i in dataset_range:
    print(i)

for i in _shuffle:
    print(i)
```

```python
repeat = tf_dataset_y.repeat(2)
list(repeat.as_numpy_iterator())
```

```python
batch = tf_dataset_y.batch(3)
for i in batch.take(5):
    print(i)
```

```python
import pathlib
import os

files_dir = r"E:\AI Projects\My Project\Files"
dir = pathlib.Path(files_dir)
print("Number of files:",len(list(dir.glob('*.*'))))

tf_files = tf.data.Dataset.list_files(str(dir/'*'), shuffle=False)
```

```python
def func(filename):
    parts = tf.strings.split(filename, os.sep)
    label = parts[-1]
    return label

t = tf_files.map((func))
for i in t:
    print(i)
```
