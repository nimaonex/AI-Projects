```python
# Eager Execution 
# code run in python
# is default
x = [[2]]
print (tf.matmul(x, x))
```

```python
# Graph Execution
# operations do not evaluate immediately
# code do not run in python
# we can force python to run graph:
@tf.function
def square(x):
    m = tf.matmul(x, x)
    return m
```

Graph benefits: 
Very fast, flexible, runs in parallel, Portable (means it is not dependent to python and you can run it every where)
Also better run in GPU, TPU

drawbacks: difficult to learn and test

Eager benefits:
pythonic interface
easy for debug
easy to prototype

drawbacks: slower, dependency to python

Best solution:

code, test and debug with eager mode and execute in graph mode

```python
def regular_func(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

graph_func = tf.function(regular_func)

x1 = tf.constant([[1,2]])
y1 = tf.constant([[2], [3]])
b1 = tf.constant(4)

reg_val = regular_func(x1, y1, b1).numpy()
graph_val = graph_func(x1, y1, b1).numpy()
assert(reg_val == graph_val)
```

```python
# runtime
import timeit
x = 10
print("eager time: ", timeit.timeit(lambda: regular_func(x1, y1, b1).numpy(), number = 10000))
print("graph time: ", timeit.timeit(lambda: graph_func(x1, y1, b1).numpy(), number = 10000))
```
